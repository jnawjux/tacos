#### (1) Add code that creates a _list of dictionaries_ called `alphabet_data`

```python
from string import (ascii_lowercase,ascii_uppercase)
```

```python
alphabet_data = []
lower = ascii_lowercase
upper = ascii_uppercase
index = range(1,27)
```

```python
for l, u, i in zip(lower, upper, index):
    temp_dict = {'lower': l, 'upper':u ,'index': i}
    alphabet_data.append(temp_dict)
```

```python
alphabet_data
```

#### (2) `import pandas as pd` and cast `alphabet_data` to a DataFrame:

```python
import pandas as pd
alphabet_df = pd.DataFrame(alphabet_data)
alphabet_df.set_index('index', inplace=True)
```
